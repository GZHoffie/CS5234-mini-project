# Experiment 1 Results

```python
import pandas as pd
import numpy as np

df_random = pd.read_csv("results/counter_matrix_l2_random.csv")
```

```python
df_zipfian = pd.read_csv("results/counter_matrix_l2_zipfian.csv")
```

```python
A_length = 5
B_length = 5

A_range = 2 ** np.arange(A_length+1)[1:]
B_range = 4 ** np.arange(B_length)

def get_A_B_Table(df,is_dependent):
    df_result = pd.DataFrame([0*np.arange(A_length)], columns=A_range, index=B_range,dtype=np.float64)
    for A in A_range:
        for B in B_range:
            rows = df[(df['A'] == A) & (df['B'] == B) & (df['is_independent'] == is_dependent)]
            if len(rows)>=1:
                df_result[A][B] = rows["error"].mean()
    return df_result

```

```python
get_A_B_Table(df_random,'dependent')
```

```python
get_A_B_Table(df_random,'independent')
```

```python
get_A_B_Table(df_zipfian,'dependent')
```

```python
get_A_B_Table(df_zipfian,'independent')
```
